## *Importing Required Libraries*


```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.cluster import KMeans
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.datasets import make_moons
```

## *Loading the Dataset*


```python
Dataset = pd.read_csv(r"E:\AI and Data Science\ML with Engr.Muhammad Javed\File\placement.csv")
```

## *Show Dataset*


```python
Dataset
```

## *Splitting Features and target Variable*


```python
x = Dataset[["cgpa"]]
y = Dataset["package"]
```

## *Splitting Dataset into Training and Testing Sets*


```python
X_train , X_test , Y_train , Y_test = train_test_split(x ,y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

## *Training Linear Regression Model and Evaluating Accuracy*


```python
lr = LinearRegression()
lr.fit(X_train , Y_train)
print(lr.score(X_train , Y_train)*100 , lr.score(X_test , Y_test)*100)
```

## *Training Decision Tree Regressor and Evaluating Performance*


```python
dt = DecisionTreeRegressor()
dt.fit(X_train , Y_train)
print(dt.score(X_train , Y_train)*100 , dt.score(X_test , Y_test)*100)
```

## *Training Support Vector Regressor (SVR) and Evaluating Performance*


```python
svr = SVR()
svr.fit(X_train , Y_train)
print(svr.score(X_train , Y_train)*100 , svr.score(X_test , Y_test)*100)
```

```python
from sklearn.ensemble import VotingRegressor
```

## *Training Voting Regressor Ensemble with Multiple Regressors*


```python
# Correct list of estimators
li = [
    ("lr1", LinearRegression()),
    ("dt1", DecisionTreeRegressor()),
    ("svr1", SVR())
]

vc = VotingRegressor(li)
vc.fit(X_train, Y_train)
```

```python
print(vc.score(X_train , Y_train)*100 , vc.score(X_test , Y_test)*100)
```

## *Creating DataFrame to Compare Predictions from Multiple Regressors with Actual Values*


```python
data = {"lr" : lr.predict(X_test) , "dt" : dt.predict(X_test) , "svr" : svr.predict(X_test) , "vc" : vc.predict(X_test) , "actual" : Y_test}

df = pd.DataFrame(data)
```
