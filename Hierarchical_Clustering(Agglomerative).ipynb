```python
import scipy.cluster.hierarchy as sc
```

## *Creating a Dendrogram for Hierarchical Clustering Visualization*


```python
plt.figure(figsize=(20 , 20))
dendrogram = sc.dendrogram(sc.linkage(Dataset , method="single" , metric="euclidean"))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.savefig("Dendrogram.png")
plt.ylabel("Euclidean Distances")
plt.show()
```

## *Creating a Dendrogram Using Complete Linkage Method for Hierarchical Clustering*


```python
plt.figure(figsize=(20 , 20))
dendrogram = sc.dendrogram(sc.linkage(Dataset , method="complete" , metric="euclidean"))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.savefig("DendrogramComplete.png")
plt.ylabel("Euclidean Distances")
plt.show()
```

## *Creating a Dendrogram Using Ward Linkage Method for Hierarchical Clustering*


```python
plt.figure(figsize=(20 , 20))
dendrogram = sc.dendrogram(sc.linkage(Dataset , method="ward" , metric="euclidean"))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.savefig("DendrogramWard.png")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

## *Applying Agglomerative Clustering with Single Linkage and Adding Cluster Labels*


```python
ac = AgglomerativeClustering(n_clusters=2 , linkage="single")
Dataset["Final Predict"] = ac.fit_predict(Dataset)

```
