## *Importing Required Libraries*


```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.cluster import KMeans
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.datasets import make_moons
```

## *Generating a Synthetic Dataset Using Make_Moons Function*


```python
x , y = make_moons(n_samples=1000 , noise=0.2)
```

## *Creating a DataFrame from Generated Features and Target Values*


```python
df = {"Feature_1" : x[:,0] , "Feature_2" : x[:,1] , "Target" : y}
data = pd.DataFrame(df)

```

## *Splitting Features and target Variable*


```python
X = data[["Feature_1" , "Feature_2"]]
Y = data["Target"]
```

## *Splitting Dataset into Training and Testing Sets*


```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , random_state=42)
```

## *Visualizing Data Points with Class Separation Using Scatter Plot*


```python
sn.scatterplot(x=X["Feature_1"] , y=X["Feature_2"] , hue=Y , data = data )
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

```

## *Training a Bagging Classifier Using Support Vector Classifier as Base Estimator*


```python
bg = BaggingClassifier(estimator=SVC() , n_estimators=30 , random_state=42)
bg.fit(X_train , Y_train)
```

## *Evaluating Bagging Classifier Accuracy on Training and Testing Data*


```python
bg.score(X_train , Y_train)*100 , bg.score(X_test , Y_test)*100
```

## *Training and Evaluating Random Forest Classifier*


```python
rf = RandomForestClassifier(n_estimators=100 , random_state=42)
rf.fit(X_train , Y_train)
rf.score(X_train , Y_train)*100 , rf.score(X_test , Y_test)*100
```
